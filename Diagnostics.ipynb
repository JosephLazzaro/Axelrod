# Running Axelrod Tounraments
In this notebook I will go through and run both the original axelrod tounraments, as well as my own tounraments with my own strategy "jlazz" as a participant.

```python
import axelrod as axl
```

```python
players = [s() for s in axl.demo_strategies]  # Create players
players.append(axl.jlazz())
players.append(axl.ZDExtort2(0.111,0.5))
tournament = axl.Tournament(players, turns=400,
      repetitions=5,
      seed=1,
 )  # Create a tournament
results = tournament.play()  # Play the tournament
results.ranked_names
```

```python
plot = axl.Plot(results)
p = plot.boxplot()
```

```python
p = plot.winplot()
```

```python
p = plot.payoff()
```
